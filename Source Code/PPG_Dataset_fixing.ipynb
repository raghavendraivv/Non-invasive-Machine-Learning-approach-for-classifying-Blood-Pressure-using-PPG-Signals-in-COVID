```python
dffeaturessignal=pd.read_csv('fixed_length_features.csv')# reading the fixed length ppg signal features file
```

```python
dffeaturessignal.drop(dffeaturessignal.columns[[0]], axis=1, inplace=True)
```

```python
dffeaturessignal.head()
```

```python
ppg_combained=pd.concat([ppg_combained,dffeaturessignal],axis=1)#Concatenating the Ppg bp dataset with fixed length feature dataframe
```

```python
ppg_combained.head()
```

```python
ppg_combained.tail()
```

```python
ppg_combained.loc[:,'0':'840']#Looking at the signal feature columns
```

```python
dropindex=[]#variable for storing the indexes to be dropped
```

```python
for i in range(len(ppg_combained)):
    rowarr=ppg_combained.loc[i,'0':'840']#Storing the signal feature columns
    if sum(rowarr)==0:
        dropindex.append(i)   #Add to dropindex if the sum of rowarr is zero
    
```

```python
dropindex #List of indexes to be dropped
```

```python
len(dropindex)
```

```python
ppg_combained.loc[dropindex]#Interpreting at the signals to be dropped
```

```python
ppg_combained.drop(dropindex,inplace=True)#Dropping the indexes present in dropindex list
```

```python
ppg_combained.head()
```

```python
#Storing the resulting dataframe obtained after dropping the dropindex indexes
ppg_combained.to_csv('PPGObtained.csv',index=False)
```
