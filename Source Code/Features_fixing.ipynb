```python
import pandas as pd
import numpy as np
```

```python
# reading first interval peak data
first_interval_peaks=pd.read_csv('first_interval_peaks.csv')
```

```python
first_interval_peaks.head()
```

```python

currmax=-float('inf')# storing the maximum amplitude of a signal
currmaxrow=0 #Storing the row of the maximum amplitude
for i in range(len(first_interval_peaks)):
  rowmax=max(first_interval_peaks.loc[i])
  if rowmax>currmax:
    currmax=rowmax
    currmaxrow=first_interval_peaks.loc[i]
print(currmax,currmaxrow)
```

```python
amp_max=pd.DataFrame(currmaxrow)#Creating a dataframe for the maxamplitude signal
```

```python
sum(amp_max.isnull().sum(axis=1))#Checking the count of null values
```

```python
num_not_null_values=len(amp_max)-sum(amp_max.isnull().sum(axis=1))#Count of not null values
num_not_null_values
```

```python
maxnotnull=num_not_null_values #Storing the not noll values count in the maxnotnull variable
```

```python
#Making the intervals length equal to maxnotnull 
first_interval_peaks=first_interval_peaks.iloc[:,:maxnotnull]
second_interval_peaks=second_interval_peaks.iloc[:,:maxnotnull]
third_interval_peaks=second_interval_peaks.iloc[:,:maxnotnull]
```
