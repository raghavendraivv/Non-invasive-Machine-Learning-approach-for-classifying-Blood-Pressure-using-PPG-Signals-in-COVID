```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,4))
sns.countplot(x="Age(year)",data=df)
```

```python
sns.distplot(df['Age(year)'],kde=False,bins=40)
```

```python
plt.figure(figsize=(12,4))
sns.countplot(x='Hypertension',hue='Sex(M/F)',data=df,palette='RdBu_r')
```

```python
continuous_features=['Age(year)','Height(cm)','Weight(kg)','Heart Rate(b/m)','BMI(kg/m^2)']
```

```python
for feature in continuous_features:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_features:
    data=df.copy()
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```
